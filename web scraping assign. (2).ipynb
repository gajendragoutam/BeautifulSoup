```python
# Q 4)  Write a python program to scrape first 10 product details which include product name , price , Image URL from
#        https://www.bewakoof.com/bestseller?sort=popular .
```

```python
url = ('https://www.bewakoof.com/bestseller?sort=popular')
web = requests.get(url)
soup = BeautifulSoup(web.text, 'html.parser')
```

```python
product = soup.find_all('div', class_='productCardBox', limit=10)
```

```python
product_details = []
for pct in product:
    name = pct.find('h3').text.strip()if pct.find('h3') else "No Name"
    price = pct.find('b').text.strip()if pct.find('b') else "No Price"
    image = pct.find('img')['src']if pct.find('img') else "No Image URL"

    product_details.append({
        'name': name,
        'price': price,
        'image_url': image
     })
```

```python
for index in range(len(product_details)):
    product = product_details[index]
    print(f"Product {index + 1}:")
    print(f"Name: {product['name']}")
    print(f"Price: {product['price']}")
    print(f"Image URL: {product['image_url']}")
    print()
```

```python
# Q 5 '' Please visit https://www.cnbc.com/world/?region=world and scrap-
#   a) headings
#   b) date
#  c) News link
```

```python
url = "https://www.cnbc.com/world/?region=world"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
articles = soup.find_all('div', class_='Card-titleContainer')
```

```python
for article in articles:
        title_f= article.find('a', class_='Card-title')
        title = title_f.get_text(strip=True)
        link = title_f["href"]
        
        print(f"Title: {title}")
        print(f"link: {link}")
        print( "date is not available")
        print()
```

```python
# Q 6 ) Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded articles/ and scrap
#  a) Paper title
#  b) date
#  c) Author
```

```python
url = ("https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/")
print(url)
```

```python
web1 = requests.get(url)
```

```python
soup= BeautifulSoup(web1.text,'html.parser')
```

```python
ar = soup.find_all('div', class_="article-listing")
```

```python

 for article in ar:
        title = article.find('h2', class_="article-title").get_text(strip=True)
        date = article.find('p', class_="article-date").get_text(strip=True)
        author = article.find('p', class_='article-authors').get_text(strip=True)
        
        print(f"Title: {title}")
        print(f"Date: {date}")
        print(f"Author: {author}")
        print()
        print()
```
