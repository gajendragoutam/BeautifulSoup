```python
url=("https://www.wikipedia.org/")
requests = requests.get(url)
```

```python
soup1 = BS(requests.content, 'html.parser')
soup1
```

```python
header_tags = []
for i in range (1,7):
    header_tags.extend(soup1.find_all(f'h{i}'))
```

```python
print("Header tags found on the webpage:")
for tag in header_tags:
    print(f"{tag.name}: {tag.get_text(strip=True)}")
```
